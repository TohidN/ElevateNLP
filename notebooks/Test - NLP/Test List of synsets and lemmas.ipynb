# Lemma methods:

Synonyms are available under `synset.lemmas()` and each resulting lemma has the following methods.

* antonyms()
* derivationally_related_forms()
* pertainyms()


```python
# derivationally related forms
for synset in wn.all_synsets():
    if synset.lemmas():
        for lemma in synset.lemmas():
            lemma_name = lemma.name().replace("_", " ")
            if lemma.derivationally_related_forms():
                print(lemma_name)
                for derivative in lemma.derivationally_related_forms():
                    derivative_name = derivative.name().replace("_", " ")
                    print(
                        f"{derivative_name} \t {derivative}:{derivative.synset().definition()}"
                    )
                print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
```

```python
# List pertainyms: Relational adjectives ("pertainyms") point to the nouns they are derived from (criminal-crime).
for synset in wn.all_synsets():
    if synset.lemmas():
        for lemma in synset.lemmas():
            lemma_name = lemma.name().replace("_", " ")
            if lemma.pertainyms():
                print(f"{lemma_name}\t {lemma}:{lemma.synset().definition()}")
                for pertainym in lemma.pertainyms():
                    pertainym_name = pertainym.name().replace("_", " ")
                    print(
                        f"{pertainym_name} \t {pertainym}:{pertainym.synset().definition()}"
                    )

                print("------------------")
```

```python
# List synsets - print only multi-word synsets
for synset in wn.all_synsets():
    if synset.lemmas():
        for lemma in synset.lemmas():
            lemma_name = lemma.name().replace("_", " ")
            words = re.split(r"\s+|-|/", lemma_name)
            if len(words) > 1:
                print(lemma)
                print(lemma_name)
                print(words)
                print("-----------")
```

```python
# List synsets - print words with (_) in their lemma
for synset in wn.all_synsets():
    if synset.lemmas():
        for lemma in synset.lemmas():
            if lemma.name().find("_") != -1:
                lemma_name = lemma.name().replace("_", " ")
                words = re.split(r"\s+|-|/", lemma_name)
                print(lemma)
                print(lemma.name())
                print(lemma_name)
                print(words)
                print("-----------")
```

```python
# List synsets - print words with (-) in their lemma
for synset in wn.all_synsets():
    if synset.lemmas():
        for lemma in synset.lemmas():
            if lemma.name().find("-") != -1:
                lemma_name = lemma.name().replace("_", " ")
                words = re.split(r"\s+|-|/", lemma_name)
                print(lemma)
                print(lemma.name())
                print(lemma_name)
                print(words)
                print("-----------")
```

```python
# list synsets which have instances
for synset in wn.all_synsets():
    # for synset in wn.synsets("good"):
    if synset.instance_hyponyms() or synset.instance_hypernyms():
        print(synset)
        # print(synset.definition())
        if synset.instance_hyponyms():
            print("> instance_hyponyms")
            print(synset.instance_hyponyms())
        if synset.instance_hypernyms():
            print("> instance_hypernyms")
            print(synset.instance_hypernyms())
        print("---------------------------------")
```

```python
# Check Attributes
a = {}
n = {}
for synset in wn.all_synsets():
    if synset.attributes() and synset.pos() == "a":
        for rel in synset.attributes():
            if rel.name() in a:
                a[rel.name()].append(synset.name())
            else:
                a[rel.name()] = [synset.name()]
    if synset.attributes() and synset.pos() == "n":
        for rel in synset.attributes():
            if synset.name() in n:
                n[synset.name()].append(rel.name())
            else:
                n[synset.name()] = [rel.name()]
# "n" for list of nouns, "a" for list of adjectives
for key in n:
    print(key)
    print(n[key])
```

```python
# SEMANTIC-NGRAM: multi-words which create a new meaning
for synset in wn.all_synsets():
    if synset.lemmas():
        for lemma in synset.lemmas():
            lemma_name = lemma.name().replace("_", " ")
            words = re.split(r"\s+|-|/", lemma_name.lower())
            if len(words) > 1:
                print(lemma_name)
                print(words)
                print("----------------")
```
