```python
# Required packages
from nltk.corpus import wordnet as wn

# for lemmatizition
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
# for parsing
import re
```

```python
# search for synsets with specific word
for synset in wn.synsets("dog"):
    print(synset)

# loop all synsets
for synset in wn.all_synsets():
    print(synset)
```

```python
# print only multi-word synsets
for synset in wn.all_synsets():
    if synset.lemmas():
        for lemma in synset.lemmas():
            lemma_name = lemma.name().replace("_", " ")
            words = re.split(r"\s+|-|/", lemma_name)
            if len(words) > 1:
                print(lemma)
                print(lemma_name)
                print(words)
                print("-----------")
```

```python
# print words with (_) in their lemma
for synset in wn.all_synsets():
    if synset.lemmas():
        for lemma in synset.lemmas():
            if lemma.name().find("_") != -1:
                lemma_name = lemma.name().replace("_", " ")
                words = re.split(r"\s+|-|/", lemma_name)
                print(lemma)
                print(lemma.name())
                print(lemma_name)
                print(words)
                print("-----------")
```

```python
# print words with (-) in their lemma
for synset in wn.all_synsets():
    if synset.lemmas():
        for lemma in synset.lemmas():
            if lemma.name().find("-") != -1:
                lemma_name = lemma.name().replace("_", " ")
                words = re.split(r"\s+|-|/", lemma_name)
                print(lemma)
                print(lemma.name())
                print(lemma_name)
                print(words)
                print("-----------")
```

```python
# list synsets which have instances
for synset in wn.all_synsets():
    # for synset in wn.synsets("good"):
    if synset.instance_hyponyms() or synset.instance_hypernyms():
        print(synset)
        # print(synset.definition())
        if synset.instance_hyponyms():
            print("> instance_hyponyms")
            print(synset.instance_hyponyms())
        if synset.instance_hypernyms():
            print("> instance_hypernyms")
            print(synset.instance_hypernyms())
        print("---------------------------------")
```

```python
# Check Attributes
a = {}
n = {}
for synset in wn.all_synsets():
    if synset.attributes() and synset.pos() == "a":
        for rel in synset.attributes():
            if rel.name() in a:
                a[rel.name()].append(synset.name())
            else:
                a[rel.name()] = [synset.name()]
    if synset.attributes() and synset.pos() == "n":
        for rel in synset.attributes():
            if synset.name() in n:
                n[synset.name()].append(rel.name())
            else:
                n[synset.name()] = [rel.name()]
# "n" for list of nouns, "a" for list of adjectives
for key in n:
    print(key)
    print(n[key])
```

```python
# list synsets's lemma pertainyms
for synset in wn.all_synsets():
    if synset.lemmas():
        for lemma in synset.lemmas():
            lemma_name = lemma.name().replace("_", " ")
            if lemma.pertainyms():
                print(lemma_name)
                print(lemma.pertainyms())
                print("------------------")
```

```python
# List all instances(Named Entities)
for synset in wn.all_synsets():
    if synset.instance_hyponyms():
        if synset.instance_hyponyms():
            print(synset)
            print(synset.definition())
            print("---------------------------------")
            for instance in synset.instance_hyponyms():
                print(instance, instance.definition())
                print(instance.hypernyms())
                print("_________________")
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
```
